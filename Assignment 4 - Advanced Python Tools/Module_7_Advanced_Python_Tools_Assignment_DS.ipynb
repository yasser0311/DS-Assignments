Assignment 1: NumPy


```python
import numpy as np

### 1. 1D Array Operations
# Create array
arr1d = np.arange(1, 21)
print("1D Array:", arr1d)

# Calculations
print("\n1a. Calculations:")
print("Sum:", np.sum(arr1d))
print("Mean:", np.mean(arr1d))
print("Median:", np.median(arr1d))
print("Standard Deviation:", np.std(arr1d))

# Indices of elements > 10
indices = np.where(arr1d > 10)
print("\n1b. Indices of elements > 10:", indices[0])

### 2. 2D Array Operations
# Create 4x4 array
arr2d = np.arange(1, 17).reshape(4, 4)
print("\n2. 4x4 Array:\n", arr2d)

# Transpose
print("\n2a. Transpose:\n", arr2d.T)

# Row-wise and column-wise sums
print("\n2b. Row sums:", np.sum(arr2d, axis=1))
print("Column sums:", np.sum(arr2d, axis=0))

### 3. 3x3 Array Operations
# Create two random arrays
arr1 = np.random.randint(1, 21, size=(3, 3))
arr2 = np.random.randint(1, 21, size=(3, 3))
print("\n3. Array 1:\n", arr1)
print("\nArray 2:\n", arr2)

# Element-wise operations
print("\n3a. Element-wise Addition:\n", arr1 + arr2)
print("\nElement-wise Subtraction:\n", arr1 - arr2)
print("\nElement-wise Multiplication:\n", arr1 * arr2)

# Dot product
print("\n3b. Dot Product:\n", np.dot(arr1, arr2))

### 4. Reshaping and Slicing
# Create and reshape array
arr = np.arange(1, 13)
reshaped = arr.reshape(3, 4)
print("\n4. Reshaped Array (3x4):\n", reshaped)

# Slice first two rows and last two columns
sliced = reshaped[:2, -2:]
print("\nSliced Array (first two rows, last two columns):\n", sliced)
```

Assignment 2: Pandas

```python
import pandas as pd

### 1. Create DataFrame
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'Diana', 'Eve'],
    'Age': [24, 27, 22, 32, 29],
    'Department': ['HR', 'Finance', 'IT', 'Marketing', 'HR'],
    'Salary': [45000, 54000, 50000, 62000, 47000]
}

df = pd.DataFrame(data)
print("1. DataFrame:\n", df)

# First five rows
print("\n1a. First five rows:\n", df.head())

# Summary statistics
print("\n1b. Summary statistics:\n", df[['Age', 'Salary']].describe())

# Average HR salary
hr_avg = df[df['Department'] == 'HR']['Salary'].mean()
print("\n1c. Average HR salary:", hr_avg)

### 2. Add Bonus column
df['Bonus'] = df['Salary'] * 0.10
print("\n2. DataFrame with Bonus:\n", df)

### 3. Filter employees aged 25-30
filtered = df[(df['Age'] >= 25) & (df['Age'] <= 30)]
print("\n3. Employees aged 25-30:\n", filtered)

### 4. Group by Department
dept_avg = df.groupby('Department')['Salary'].mean()
print("\n4. Average salary by department:\n", dept_avg)

### 5. Sort and save to CSV
sorted_df = df.sort_values('Salary')
sorted_df.to_csv('sorted_salaries.csv', index=False)
print("\n5. DataFrame sorted by salary and saved to CSV.")
```

Assignment 3: Matplotlib

```python
import matplotlib.pyplot as plt

### 1. Line Plot
x = [1, 2, 3, 4, 5]
y = [10, 15, 25, 30, 50]

plt.figure(figsize=(8, 5))
plt.plot(x, y, marker='o', linestyle='-', color='blue')
plt.title('Simple Line Plot')
plt.xlabel('X Values')
plt.ylabel('Y Values')
plt.grid(True)
plt.show()

### 2. Bar Graph
students = ['John', 'Jane', 'Alice', 'Bob']
marks = [75, 85, 60, 90]

plt.figure(figsize=(8, 5))
colors = ['skyblue', 'lightgreen', 'salmon', 'gold']
plt.bar(students, marks, color=colors)
plt.title('Student Marks')
plt.xlabel('Students')
plt.ylabel('Marks')
plt.show()

### 3. Pie Chart
regions = ['North America', 'Europe', 'Asia', 'Others']
revenue = [45, 25, 20, 10]
explode = (0.1, 0, 0, 0)  # Highlight largest slice

plt.figure(figsize=(8, 5))
plt.pie(revenue, labels=regions, autopct='%1.1f%%',
        explode=explode, shadow=True)
plt.title('Revenue Distribution by Region')
plt.show()

### 4. Histogram
data = np.random.randint(1, 101, size=1000)

plt.figure(figsize=(8, 5))
plt.hist(data, bins=20, edgecolor='black', color='purple')
plt.title('Frequency Distribution of Random Integers (1-100)')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```
